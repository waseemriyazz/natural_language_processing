```python
df['num_words'] = pd.to_numeric(df['num_words'], errors='coerce')  # 'coerce' will turn invalid parsing into NaNs

# Check the data type after conversion
print(df['num_words'].dtype)
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Define the bin ranges
mean = np.mean(df['num_words'].to_numpy())
std_dev = np.std(df['num_words'])

# Define the bin edges
bins = np.linspace(min(df['num_words']), max(df['num_words']), 35)

# Plotting
plt.figure(figsize=(10, 6))
plt.hist(df['num_words'], bins=bins, color='skyblue', edgecolor='black')

# Adding labels and title
plt.title('Distribution of Number of Words')
plt.xlabel('Number of Words')
plt.ylabel('Frequency')

# Customizing x-axis labels
plt.xticks(bins, rotation=45)



plt.grid(True)
plt.show()
```

```python


# Get the top 3 authors and their message counts
top_3_authors = df['author_name'].value_counts().head(3)

# Extract the names, emails, and number of posts of the top 3 authors
top_3_names = top_3_authors.index.tolist()
top_3_emails = [df[df['author_name'] == author]['email_of_author'].iloc[0] for author in top_3_names]
top_3_posts = top_3_authors.tolist()

# Print the results
for i in range(3):
    print(f"Name: {top_3_names[i]}, Email: {top_3_emails[i]}, Number of Posts: {top_3_posts[i]}")
```

```python
# Find the user with the most number of positive articles
positive_counts = df[df['sentiment_text_blob'] == 'POSITIVE']['author_name'].value_counts()
most_positive_user = positive_counts.idxmax()
num_positive_articles = positive_counts.max()

# Find the user with the most number of negative articles
negative_counts = df[df['sentiment_text_blob'] == 'NEGATIVE']['author_name'].value_counts()
most_negative_user = negative_counts.idxmax()
num_negative_articles = negative_counts.max()

# Print the results
print(f"User with most number of positive articles: {most_positive_user}, Number of positive articles: {num_positive_articles}")
print(f"User with most number of negative articles: {most_negative_user}, Number of negative articles: {num_negative_articles}")
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Convert the 'tokenized_text' column to a string
df.dropna(inplace=True)
all_text = ' '.join(df['tokenized_text'].astype(str))

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count occurrences of each topic
topic_counts = df['dominant_topic'].value_counts()

# Plot the graph
plt.figure(figsize=(12, 6))
topic_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Articles for Each Dominant Topic')
plt.xlabel('Dominant Topic')
plt.ylabel('Number of Articles')
plt.xticks(rotation=45, ha='right')  # Adjust rotation angle and alignment of x-axis labels
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count the occurrences of each sentiment
sentiment_counts = df['sentiment_text_blob'].value_counts()

# Plotting
plt.figure(figsize=(8, 8))
plt.pie(sentiment_counts, labels=None, autopct='%1.1f%%', startangle=140, colors=['green', 'red', 'blue'])

# Adding title
plt.title('Distribution of Sentiments')

# Show legend
plt.legend(sentiment_counts.index, title='Sentiments')

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Show plot
plt.show()

```
